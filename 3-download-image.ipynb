```python
import pandas as pd
import numpy as np
from pathlib import Path
from multiprocessing import Pool
from crawler import Crawler
crawler = Crawler()
data = pd.read_csv('chinese-celebrity-faces.csv')
images = data['IMAGE_PATH'].unique()
downloaded_images = list(Path('./images').glob("*"))
data['downloaded'] = data['IMAGE_PATH'].isin(downloaded_images)

thread = 2

print('DOWNLOAD FACE IMAGES')
print('=' * 60)
print(f'* total {len(images)} images')
print(f'* downloaded {len(downloaded_images)} images')
#print(f'* pending {len(pending_images)} images')
print('=' * 60)
```

```python
def download(args):
    return crawler.save_poster(**args)
```

```python
urls = [
    {"link": "https://image.ijq.tv/201608/09/17-45-52-27-26.jpg",
     "image_path": "images/17-45-52-27-26.jpg"},
    {"link": "https://image.ijq.tv/201909/25/14-13-18-14-47.png",
     "image_path": "images/14-13-18-14-47.png"}
]
```

```python
with Pool(2) as pool:
    res = pool.imap(download, urls)
```
